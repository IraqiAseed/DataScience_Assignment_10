```python
#1) Using titanic_full.csv dataset, 
#extract Mr,Miss,Mrs,etc.. from 'Name' column and use it to create new column and call it 'Title', 
#then plot a countplot 
#showing the count for each title in the dataset.
```

```python
def extract_title(name):
    title = name.split(",")[1].split(".")[0]
    return title
df['Title'] = df['Name'].apply(extract_title)
df
```

```python
chart = sns.countplot(x='Title',data=df)
chart.set_xticklabels(
    chart.get_xticklabels(), 
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='x-large')
None #don't show the label objects
```

```python
#Using visitor-interests.csv dataset, 
#from the first 50 row extract Browser, OS, Device data from 'UserAgent' column 
#and use it to create new columns for Browser, OS, Device.
```

```python
df_visitor_interests = pd.read_csv('visitor-interests.csv')
df_visitor_interests
```

```python
df_visitor_interests_fifty = df_visitor_interests.loc[:50]
df_visitor_interests_fifty
```

```python
import user_agents

def extract_web_data(ua):
    ua = user_agents.parse(ua)
    return pd.Series((ua.os.family, ua.browser.family, ua.device.family))

df_visitor_interests_fifty[['OS', 'Browser', 'Device']] = df_visitor_interests_fifty['UserAgent'].apply(extract_web_data)
df_visitor_interests_fifty
```

```python
#Using visitor-interests.csv dataset, from the first 50 row extract latitude, longitude from 'IP' column.
```

```python
from ip2geotools.databases.noncommercial import DbIpCity as ip2geo
def extract_latitude_longitude_from_ip(ip):
    try:
        response = ip2geo.get(ip, api_key='free')
        return pd.Series((response.latitude,response.longitude))
    except:
        return np.NaN
    
    
df_visitor_interests_fifty[['latitude', 'longitude']] = df_visitor_interests_fifty['IP'].apply(extract_latitude_longitude_from_ip)
df_visitor_interests_fifty
```

```python
#Using visitor-interests.csv dataset, 
#from the first 50 row extract latitude, longitude from 'IP' column 
#and use it to create new 'Dist_From_Egypt' column calculating distance with Egypt Lat/Lonng, 
#you can get this location from google.
```

```python
from datasist.feature_engineering import haversine_distance
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="Data Science Assignment")
Egypt_location = geolocator.geocode("Egypt") #there is no exact location for a country .. !!

#lat,long
df_visitor_interests_fifty['Dist_From_Egypt'] = haversine_distance(df_visitor_interests_fifty['latitude'],df_visitor_interests_fifty['longitude'],Egypt_location.latitude,Egypt_location.longitude)
df_visitor_interests_fifty
```
